```python
def quantile_score_2(q,y,f):
    if f>y:
        score = 2*(1-q)*(f-y)
    else:
        score = 2*q*(y-f) 
    return score      

def s(y,q):
    return y

def h(y,q): 
    return (-y)*q

def one(y,f):
    if y<=f:
        return 1
    else:
        return 0

def quantile_score(q,y,f,scale=2):
    return abs(scale*(q*s(f,q)+(s(y,q)-s(f,q))*one(y,f)+h(y,q)))
```
